```
from fastai.vision import *
```

## Dowloading images for coloseum rome, eifel tower and UT tower

```
from google.colab import drive
drive.mount('/content/gdrive', force_remount=True)
root_dir = "/content/gdrive/My Drive/"
base_dir = root_dir + 'fastai-v3/'
```

```
path = Path(base_dir + 'data/landmarks')
```

```
folder = 'coloseum'
file = 'urls_colosseum_rome.txt'
dest = path/folder
dest.mkdir(parents=True, exist_ok=True)
download_images(path/file, dest, max_pics=500)
```

```
folder = 'eiffel'
file = 'urls_eiffel_tower_paris.txt'
dest = path/folder
dest.mkdir(parents=True, exist_ok=True)
download_images(path/file, dest, max_pics=500)
```

```
folder = 'uttower'
file = 'urls_ut_tower.txt'
dest = path/folder
dest.mkdir(parents=True, exist_ok=True)
download_images(path/file, dest, max_pics=500)
```

```
classes = ['coloseum','eiffel','uttower']
```

```
#removing all bad images
for c in classes:
    print(c)
    verify_images(path/c, delete=True, max_size=500)
```

## Loading data

```
np.random.seed(42)
data = ImageDataBunch.from_folder(path, train=".", valid_pct=0.2,
        ds_tfms=get_transforms(), size=224, num_workers=4).normalize(imagenet_stats)
```

```
data.classes
```

```
data.show_batch(rows=4, figsize=(7,8))
```

```
data.classes, data.c, len(data.train_ds), len(data.valid_ds)
```

## Training model using ResNet34

```
learn = create_cnn(data, models.resnet34, metrics=error_rate)
```

```
# training 4 epochs usng one_cycle method
learn.fit_one_cycle(4)
```

```
learn.save('stage-1')
```

## Fine tuning ResNet34

```
learn.load('stage-1')
```

```
#Using learning rate finder to identify the learning rate at which the conv net can train the fastest
learn.unfreeze()
learn.lr_find()
```

```
# Loss diverges ater 5 epochs
learn.recorder.plot()
```

```
#unfreezing all layers
learn.unfreeze()
```

```
#passing range of learning rates based on the plot
learn.fit_one_cycle(5, max_lr=slice(1e-6,1e-5))
```

```
learn.save('stage-2')
```

## Interpretation of ResNet34


```
learn.load('stage-2');
```

```
interp = ClassificationInterpretation.from_learner(learn)

losses,idxs = interp.top_losses()

len(data.valid_ds)==len(losses)==len(idxs)
```

```
interp.plot_confusion_matrix()
```

```
interp.plot_top_losses(9, figsize=(15,11))
```

Looks like the most confused are images that are arbitary images that do not correspond to any particular landmark.

```
interp.most_confused(min_val=2)
```

## Data Augmentation

```
np.random.seed(42)
data = ImageDataBunch.from_folder(path, train=".", valid_pct=0.9, bs=32, 
        ds_tfms=get_transforms(do_flip=True, max_rotate=0, max_zoom=1, max_lighting=0, max_warp=0
                              ),size=224, num_workers=4).normalize(imagenet_stats)
```

```
data.show_batch(rows=4, figsize=(7,8))
```

## Trying ResNet50

```
learn = create_cnn(data, models.resnet50, metrics=error_rate)
```

```
learn.fit_one_cycle(5)
```

```
learn.save('stage-1-50')
```

```
learn.unfreeze()
```

```
learn.lr_find()
```

```
learn.recorder.plot()
```

```
learn.fit_one_cycle(3,slice(5e-4,1e-3))
```
